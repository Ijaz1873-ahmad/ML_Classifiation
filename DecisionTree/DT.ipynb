```python
# import libraries
import sys
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
fileURL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/drug200.csv'
```

```python
df = pd.read_csv(fileURL, delimiter=",")
df.head()
```

```python
df.isna().sum()
```

```python
# preprocessing 
X = df[["Age", "Sex", "BP", "Cholesterol", "Na_to_K"]].values
X[:5]
```

```python
from sklearn.preprocessing import LabelEncoder
leSex = LabelEncoder().fit(['F','M'])
X[:, 1] = leSex.transform(X[:, 1])

X[:3]
```

```python
le_BP = LabelEncoder().fit([ 'LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])
X[:3]
```

```python
le_Chol = LabelEncoder().fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3]) 

X[0:3]
```

```python
y = df["Drug"]
y[:5]
```

```python
from sklearn.model_selection import train_test_split

X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)

drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
print(drugTree) 

drugTree.fit(X_trainset,y_trainset)

# predictions...
predTree = drugTree.predict(X_testset)
```

```python
# evaluation
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```
